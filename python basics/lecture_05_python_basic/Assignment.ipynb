Take a character from the user(use if and match).
Check if it is a vowel (a, e, i, o, u) or not.

````python
ch = input( ).lower()

if ch in ['a', 'e', 'i', 'o', 'u']:
    print("Vowel")
else:
    print("Not a vowel")
````

````python
ch = input("Enter a character: ").lower()

match ch:
    case 'a' | 'e' | 'i' | 'o' | 'u':
        print("Vowel")
    case _:
        print("Not a vowel")
````

Q9*
Given a list of numbers:

if number is even → add it to even_list

otherwise → add it to odd_list
Print both lists at the end.

````python
even_l=[]
odd_l=[]
x=int(input())
for i in range(1,x+1):
  y=int(input())
  if y%2==0:
    even_l.append(y)
  else:
    odd_l.append(y)

print(even_l)
print(odd_l)
````

Q10.
Take a list of words.
Count how many words have length > 5.

````python
li=list(input().split())
count=0
for i in li:
  if len(i)>5:
    count+=1
  else:
    continue
print(count)
````

Write a program that:

Checks if a number is prime using a for loop

````python
x=int(input())
if x==1:
  print("not prime")
else:
  is_prime = True
  for i in range(2,x+1):
    if x%i==0:
      is_prime=False
      break
    else :
      is_prime=True

if(is_prime):
  print("prime")
else:
   print("not prime")
````
