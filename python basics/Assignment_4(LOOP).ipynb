Given a number N. Print the summation of all numbers between 1 and N inclusive.

# Triangle Star Pattern

Input: 4
Output:
* ## *
* ## **
* ## ***
* ## ****


```python
x=int(input("enter a number :"))
for i in range(1,x+1):
  for j in range(i):
    print("*", end="")
  print()
```

# Count Vowels in a String
## Take a string input and count how many vowels it contains.

```python
name=list(input("enter a string please : " ).lower())
count_a=count_e=count_u=count_i=count_o=0
print(name)
for i in name:
  if i=="a":
    count_a+=1
  if i=="e":
    count_e+=1
  if i=="o":
    count_o+=1
  if i=="u":
    count_u+=1
  if i=="i":
    count_i+=1
print(f"the a is {count_a} \n the i is {count_i} \n the u is {count_u} \n the e is {count_e} \n the o is {count_o}   ")
```

Given a list of  numbers. Count how many of these values are even, odd, positive and negative.

l=-5 0 -3 -4 12

Output:

Even: 3

Odd: 2

Positive: 1

Negative: 3


```python
li=list(map(int , input("enter numbers :").split()))
count_p=count_n=0
count_even=count_odd=0
for i in li:
  if i %2==0:
    count_even+=1
  else:
    count_odd+=1

  if i>0:
    count_p+=1
  else :
    count_n+=1
print(f"Even numbers: {count_even}")
print(f"Odd numbers: {count_odd}")
print(f"Positive numbers: {count_p}")
print(f"Negative numbers: {count_n}")
```
